***CONTROL STRUCTURES***

In programming, control structures allow modifying the flow of execution of the instructions in a program.

With control structures you can:

-According to a condition, execute a group or another of statements (If-Then-Else)
-According to the value of a variable, execute a group or another of statements (Select-Case)
-Execute a group of statements while a condition is met (Do-While)
-Execute a group of statements until a condition is met (Do-Until)
-Execute a group of statements a specified number of times (For-Next)

***Conditional statements: if***

The if ... control structure allows a program to execute instructions when a condition is met. In English "if" means "if" (condition).

Syntax of the conditional statement if ...
The syntax of the if construct is as follows:

if condition:
     here are the orders that are executed if the condition is true
     and that can occupy several lines
The execution of this construction is as follows:

The condition is always evaluated.
If the result is True, the block of statements is executed
If the result is False, the statement block is not executed.

```python
numero = int(input("Escriba un número positivo: "))
if numero < 0:
    print("¡Le he dicho que escriba un número positivo!")
print(f"Ha escrito el número {numero}")
```

***Branches: if / else***

The if ... else ... control structure allows a program to execute some instructions when a condition is met and other instructions when that condition is not met. In English "if" means "if" (condition) and "else" means "if not". The command in Python is written like this:

Syntax of the conditional statement if ... else ...
The syntax of the if ... else ... construction is as follows:

if condition:
     here are the orders that are executed if the condition is true
     and that can occupy several lines
else:
     and here are the orders that are executed if the condition is
     false and that can also occupy several lines
The execution of this construction is as follows:

The condition is always evaluated.
If the result is True, only statement block 1 is executed
If the result is False, only statement block 2 is executed.

```python
edad = int(input("¿Cuántos años tiene? "))
if edad < 18:
    print("Es usted menor de edad")
else:
    print("Es usted mayor de edad")
print("¡Hasta la próxima!")
```

***Nested Sonditional Statements***

A conditional statement can itself contain another nested statement.

For example, the following program displays the color obtained by mixing two colors on the screen:

```python
print("Este programa mezcla dos colores.")
print("  r. Rojo      a. Azul")
primera = input("  Elija un color (r o a): ")
if primera == "r":
    print("  a. Azul      v. Verde")
    segunda = input("  Elija otro color (a o v): ")
    if segunda == "a":
        print("La mezcla de Rojo y Azul produce Magenta.")
    else:
        print("La mezcla Rojo y Verde produce Amarillo.")
else:
    print("  v. Verde    r. Rojo")
    segunda = input("  Elija otro color (v o r): ")
    if segunda == "v":
        print("La mezcla de Azul y Verde produce Cian.")
    else:
        print("La mezcla Azul y Rojo produce Magenta.")
print("¡Hasta la próxima!")
```

***The for Loop***

In general, a loop is a control structure that repeats a block of instructions. A for loop is a loop that repeats the statement block a predetermined number of times. The block of instructions that is repeated is usually called the body of the loop, and each repetition is usually called an iteration.

The syntax of a for loop is as follows:

for variable in iterable element (list, string, range, etc.):
     loop body
It is not necessary to define the control variable before the loop, although a variable already defined in the program can be used as the control variable.

The body of the loop is executed as many times as there are elements in the traversable element (elements of a list or of a range (), characters of a string, etc.).

```python
print("Comienzo")
for i in [0, 1, 2]:
    print("Hola ", end="")
print()
print("Final")
```

***The while Loop***

A while loop allows repeating the execution of a group of instructions as long as a condition is met (that is, as long as the condition has the value True).

The syntax of the while loop is as follows:

while condition:
     loop body
The execution of this while control structure is as follows:

Python evaluates the condition:
if the result is True, the body of the loop is executed. After the body of the loop is executed, the process is repeated (the condition is evaluated again and, if true, the body of the loop is executed again) over and over as long as the condition is true.
if the result is False, the body of the loop does not execute and the rest of the program continues execution.
The variable or variables that appear in the condition are often called control variables. Control variables must be defined before the while loop and modified in the while loop.

```python
i = 1
while i <= 3:
    print(i)
    i += 1
print("Programa terminado")
```
